```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
# iris = datasets.load_iris(as_frame=True).frame
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris
```

```python
import seaborn as sns
counts = iris["variety"].value_counts()
sns.barplot(data = counts)
```

```python
iris.info()
```

```python
iris[iris.duplicated(keep=False)]
```

```python
labelencoder = LabelEncoder()
iris.iloc[:, -1] = labelencoder.fit_transform(iris.iloc[:,-1])
iris.head()
```

```python
iris.info()
```

```python
iris['variety'] = pd.to_numeric(labelencoder.fit_transform(iris['variety']))
print(iris.info())
```

```python
iris.head(3)
```

#### observations
- There are 150 rows and 5 columns
- There is one duplicated row (101 and 142 are duplicated)
- There are no null values
- The x-columns are sepal.length,sepal.width,petal.length and petal.width
- All x columns are continuous
- There are three flower categories

## Transform the y-column to categorical using LabelEncoder

```python
X=iris.iloc[:,0:4]
Y=iris['variety']
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state = 1)
x_train
```

# observation

- the target column ("variety") is still object type

```python
model = DecisionTreeClassifier (criterion = 'entropy',max_depth = None)
model.fit(x_train,y_train)
```

```python
plt.figure(dpi=1200)
tree.plot_tree(model);
```

```python
x_train
```

```python
fn=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
cn=['setosa','versicolor','virginica']
plt.figure(dpi=1200)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```

```python
preds = model.predict(x_test)
preds
```

```python
print(classification_report(y_test,preds))
```
